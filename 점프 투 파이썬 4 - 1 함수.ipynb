### 함수란 무엇인가?

* 입력값을 가지고 어떤 일을 수행한 다음에 그 결과물을 내어놓는 것.

### 함수를 사용하는 이유

* 반복되는 부분 쉽게 작성할 수 있고, 프로그램의 흐름을 일목요연하게 볼 수 있다.

### 함수의 구조

```py
def 함수명(매개변수):
    <수행할 문장1>
    <수행할 문장2>
    ...
```

### 매개변수와 인수

* 매개변수(parameter)은 함수에 입력으로 전달된 값을 받는 **변수**, 인수(arguments)는 함수를 호출할 때 전달하는 **입력값**

예)
```py
def add(a, b):   # a, b는 매개변수
    return a+b
  
print(add(3,4))  # 3, 4는 인수
```

### 입력값과 결괏값에 따른 함수의 형태

1. 일반적인 함수 -> 매개변수 o, return 값 o  
2. 입력값이 없는 함수 -> 매개변수 x, return 값 o  
3. 결괏값이 없는 함수 -> 매개변수 o, return 값 x  
4. 입력값, 결괏값 둘 다 없는 함수 -> 매개변수 x, return 값 x

### 매개변수 지정하여 호출하기

```py
def add(a, b):
    return a + b
    
result = add(a = 3, b = 7)   # 매개변수 지정해서 사용
result = add(b = 5, a = 3)   # 순서에 상관없이 사용할 수 있다.
```

### 입력값이 몇 개가 될지 모를 때

* 가변인자일 경우 매개변수 앞에 *을 입력해주어서 사용한다.

예)
```py
def add_many(*args):
    result = 0
    for i in args:
        result = result + i
    return result
```

### 함수 안에서 함수 밖의 변수를 변경하는 방법

1. return 사용하기  
2. global 명령어 사용하기

예)

```py
a = 1 
def vartest(): 
    global a 
    a = a+1
```

여기서 global a는 함수 안에서 함수 밖의 a 변수를 직접 사용하겠다는 뜻이고, 함수는 독립적으로 존재하는 것이 좋기 때문에 외부 변수에 종속적인 global 명령어는 사용하지 않는 것이 좋다.

3. lambda

함수를 생성할 때 사용하는 예약어로 def와 동일한 역할을 한다. 보통 함수를 한줄로 간결하게 만들 때 사용한다.

lambda 매개변수1, 매개변수2,... : 매개변수를 이용한 표현식

예)

```py
add = lambda a, b: a+b
result = add(3, 4)
```
