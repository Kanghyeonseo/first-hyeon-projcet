5.	다음과 같은 기능을 가지는  TV클래스를 구현하여라. 그리고 이 클래스를 이용하여 인스턴스를 생성하여라. 이 클래스는 다음과 같은 속성과 메서드를 가진다.

<span style = "color : blue">**예제 코드**</span>

<span style = "color : blue">**실행 결과**</span>

TV가 꺼짐 상태입니다.  
볼륨 = 5, 채널 = 0  
볼륨 = 5, 채널 = 200  
볼륨 = 6, 채널 = 0  


```python
class TV:
    MIN_VOLUME = 0
    MAX_VOLUME = 20
    MIN_CHANNEL = 0
    MAX_CHANNEL = 20
    def __init__(self, volume = 5, channel = 0, is_on = False):
        self.__volume = volume
        self.__channel = channel
        self.__is_on = is_on
       
    def __str__(self):
        if self.__is_on == False:
            return "TV가 꺼짐 상태입니다."
        else:
            return "볼륨 = {}, 채널 = {}".format(self.__volume, self.__channel)
    
    def toggle_power(self):
        if self.__is_on == False:
            self.__is_on = True
        else:
            self.__is_on = False
        
    def get_channel(self):
        return self.__channel
    
    def get_volume(self):
        return self.__volume
    
    def set_channel(self, choice):
        if (choice < 0 or choice > 200):
            print("채널 오류")
        else:
            self.__channel = choice
            
    def set_volume(self, choice):
        if (choice < 0 or choice > 20):
            print("볼륨 오류")
        else:
            self.__volume = choice
    
    def volume_up(self):
        if self.__volume < TV.MAX_VOLUME:
            self.__volume += 1
    
    def voulme_down(self):
        if self.__volume > TV.MIN_VOLUME:
            self.__volume -= 1
            
    def channel_up(self):
        if self.__channel + 1 > TV.MAX_CHANNEL:
            self.__channel = TV.MIN_CHANNEL
        else:
            self.__channel += 1
            
    def channel_down(self):
        if self.__channel - 1 < TV.MIN_CHANNEL:
            self.__channel = TV.MAX_CHANNEL
        else:
            self.__channel -= 1
            
        
my_tv = TV()
print(my_tv)
my_tv.toggle_power()
print(my_tv)
my_tv.set_channel(200)
print(my_tv)
my_tv.volume_up()
my_tv.channel_up()
print(my_tv)
```
